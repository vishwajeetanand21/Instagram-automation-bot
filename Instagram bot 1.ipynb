# Problem 1 : Login to your Instagram Handle
Submit with sample username and password

```python
def Login_To_Insta(userName,Password):
    # Locating Username TextBox then enter username
    wait.until(EC.presence_of_element_located((By.NAME,'username'))).send_keys(userName)

    # Locating Password TextBox then enter password
    wait.until(EC.presence_of_element_located((By.NAME,'password'))).send_keys(Password)

    # Locating Login Button then submit it
    wait.until(EC.presence_of_element_located((By.CLASS_NAME,'_acan'))).submit()
```

# Problem 2 : Type for “food” in search
Type for “food” in search bar and print all the names of the Instagram Handles that are displayed in list after typing “food” bar and print all the names of the Instagram Handles that are displayed in list after typing “food”
Note : Make sure to avoid printing hashtags

```python
def Search(string):
    # Locating and clicking the Search button
    wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/div/a/div'))).click()
    
    
    # Locating Search TextBox then enter search string
    wait.until(EC.presence_of_element_located((By.XPATH,'//input[@class="_aauy focus-visible"]'))).send_keys(string)

    # Find all Search handle
    data=wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_abm4')))

    # Iterate over search handle and extract the text and print by ignored hashtag
    for i in data:
        result=i.text.split() #spliting the result into words
    
        if(len(result)==0): #if there is any empty result then skip
            continue
    
        if(result[0][0]=='#'): #if any name starts with '#', avoid printing hashtag
            print(result[0][1:])
        else:
            print(result[0])

    sleep(6)
    # Clearing search bar
    driver.find_element(By.CLASS_NAME,'_aawn._9-lv').click()
    
    # Click on the search button to unhide the search button
    wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div/div[2]/div[2]/div/a/div'))).click()
```

```python
# serach for food
Search('food')
```

# Problem 4 : Follow/Unfollow given handle -
Open the Instagram Handle of “So Delhi”
Start following it. Print a message if you are already following
After following, unfollow the instagram handle. Print a message if you have already unfollowed.

```python
def Follow():
    # Locating Follow Button
    Handle = wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class,"_acan")]')))

    # check already follow or not
    if Handle.text=='Follow':

        # not following so that follow it
        sleep(3)
        Handle.click()
        print("Now you are following 'So Delhi'")
    else:
        print('Already following')
```

```python
def Unfollow():
    # Locating Follow Button
    Handle = wait.until(EC.presence_of_element_located((By.XPATH,'//button[contains(@class,"_acan")]')))

    # check already follow or not
    if Handle.text=='Follow':
        print('Already unfollowed')

    else:
        # if following so that unfollow it
        Handle.click()
        sleep(5)
        wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div/div[7]/div/div/div/div/div'))).click()
        print("Now you have unfollowed 'So Delhi'")
```

# Problem 5 : Like/Unlike posts
Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.


```python
def Liking():   
    # Getting Handle of First Post then click on it
    handle=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aabd _aa8k _aanf"]')))
    handle.click()
    
    count=1 #variable to keep track of number of posts
    
    # Iterate over 30 post
    for i in range(30):

        # Locating the Like Button
        Like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class="_aamw"]/button')))

        # Getting the Text the Button
        GetText = BeautifulSoup(Like.get_attribute('innerHTML'),'html.parser').svg['aria-label']

        if GetText=='Like':
            # if not Liked then like it
            Like.click()
            print("Post number {} liked".format(count))
            count+=1
        else:
            print('Post number {} already liked'.format(count))
            count+=1

        # Locating the Next Post Button and click on it
        wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class=" _aaqg _aaqh"]/button'))).click()

    # when 30 post completed then back to the home page
    driver.back()
```

```python
def Unliking(): 
    try:
        # Getting Handle of First Post then click on it
        handle=wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aabd _aa8k _aanf"]')))
        handle.click()

        count=1 #variable to keep track of number of posts

        # Iterate over 30 post
        for i in range(30):

            # Locating the Like Button
            Like = wait.until(EC.presence_of_element_located((By.XPATH,'//span[@class="_aamw"]/button')))

            # Getting the Text the Button
            GetText = BeautifulSoup(Like.get_attribute('innerHTML'),'html.parser').svg['aria-label']

            if GetText=='Unlike':
                # if not Liked then like it
                sleep(1) #we have added some sleep because if we dislike post very fast then we will get a warning from Instagram regarding security
                Like.click()
                print("Post number {} unliked".format(count))
                count+=1
            else:
                print('Post number {} already unliked'.format(count))
                count+=1


            # Locating the Next Post Button and click on it
            wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class=" _aaqg _aaqh"]/button'))).click()

        # when 30 post completed then back to the home page
        driver.back()
    
    except ElementClickInterceptedException :
        print("Something Went Wrong! Because Instagram has made some restrictions in unliking the post")
```

# Problem 6 : Extract list of followers
Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.
Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.


```python
def Extract_list_of_followers():
    # Locating the Followers button and click on it
    wait.until(EC.presence_of_element_located((By.XPATH,'/html/body/div[2]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[2]/section/main/div/header/section/ul/li[2]/a'))).click()

    # Locating Followers List Frame
    frame = wait.until(EC.presence_of_element_located((By.CLASS_NAME,"_aano")))

    # Scrolling untill first 500 user is located
    for i in range(50):
        sleep(2)
        driver.execute_script('arguments[0].scrollTop=arguments[0].scrollHeight',frame)

    # getting the handle of all List of user
    handle = wait.until(EC.presence_of_all_elements_located((By.CLASS_NAME,'_ab8y')))

    All_User=[]
    # iterate over handle
    for i in handle[:500]:
        All_User.append(i.text.split('\n')[0])

    sleep(1)
    # Locating Cross Button and click on 
    driver.find_element(By.XPATH,'/html/body/div[2]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/button').click()

    return All_User
```

```python
# Search for Given Insta Holder Id and open his profile
Search_and_Open_Profile('so delhi')
sleep(3)

# Extract list of followers of foodtalkindia
followers_of_foodtalkindia=Extract_list_of_followers()
count=1
for i in followers_of_foodtalkindia:
    print(count,i)
    count+=1
```

# Problem 7 : Check the story of ‘coding.ninjas’.
Consider the following Scenarios and print error messages accordingly -
If You have already seen the story.
Or The user has no story.
Or View the story if not yet seen.

```python
def check_story():
    try:
        
        # Locating the story
        story = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aarf _aarg"]')))
        
        #check the Profile photo size to find out story is available or not
        Height = wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aarf _aarg"]/canvas'))).get_attribute('height')
        
        if Height=='210':
            print("Viewing the story")
            sleep(5)
            wait.until(EC.presence_of_element_located((By.XPATH,'//div[@class="_aarf _aarg"]'))).click()
            print("You have viewed the story")
            
        elif Height=='208':
            print('Already seen the story')
        
        else:
            print("No story is available to view")
    except:
        
        # Getting error it means No story is available
        print('No story is available to view')
```
